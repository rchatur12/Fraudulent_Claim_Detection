### **4.4 Perform bivariate analysis** <font color="red">[10 Marks]</font>

#### **4.4.1** Target likelihood analysis for categorical variables

```python
# Write a function to calculate and analyse the target variable likelihood for categorical features
from scipy.stats import chi2_contingency
def target_likelihood_analysis(data, target, cat_cols):
    likelihoods = {}
    baseline_fraud_rate = data[target].value_counts(normalize=True).get('Y', 0)
    print(f"Baseline fraud rate (P(fraud_reported = Y)): {baseline_fraud_rate:.3f}")

    for col in cat_cols:
        # Likelihood of fraud & non-fraud with each feature
        likelihood = data.groupby(col)[target].value_counts(normalize=True).unstack().fillna(0)
        # Frequency of each category
        freq = data[col].value_counts().sort_index()
        # Combine likelihood and frequency into a single DataFrame
        result = likelihood.join(freq.rename('Frequency')).reset_index()
        result = result.reset_index().rename(columns={col: 'Category'})
        result['Fraud_Rate_Diff'] = result['Y'] - baseline_fraud_rate

        # chi-squared test
        contingency_table = pd.crosstab(data[col], data[target])
        chi2, p_value, _, _ = chi2_contingency(contingency_table)

        # Store results
        likelihoods[col] = {
            'likelihood': result,
            'chi2': chi2,
            'p_value': p_value
        }

        # Print results
        print(f"\nAnalysis for {col}:")
        print(f"Chi-squared test: chi2 = {chi2:.2f}, p-value = {p_value:.4f}")
        print(result[['Category', 'N', 'Y', 'Frequency', 'Fraud_Rate_Diff']].to_string(index=False))

        # stacked bar plot
        plt.figure(figsize=(10, 6))
        likelihood.plot(kind='bar', stacked=True, color=['#1f77b4', '#ff7f0e'])
        plt.title(f'Fraud Likelihood by {col}')
        plt.xlabel(col)
        plt.ylabel('Proportion')
        plt.legend(title='Fraud Reported')
        plt.tight_layout()
        plt.show()

    # Print findings
    print("\nSummary of Target Likelihood Analysis:")
    for col in cat_cols:
        p_value = likelihoods[col]['p_value']
        significant = p_value < 0.05
        high_fraud = likelihoods[col]['likelihood'][likelihoods[col]['likelihood']['Y'] > baseline_fraud_rate]['index'].tolist()
        print(f"- {col}: {'Significant' if significant else 'Not significant'} (p-value = {p_value:.4f}). "
              f"Categories with higher-than-baseline fraud rate: {high_fraud if high_fraud else 'None'}.")

    return likelihoods


#select catagorical columns only
categorical_cols = X_train.select_dtypes(include=['object']).columns
#call function to do target variable likelihood
likelihoods = target_likelihood_analysis(pd.concat([X_train, y_train], axis=1), 'fraud_reported', categorical_cols)
```

### **6.4 Combine values in Categorical Columns** <font color="red">[6 Marks]</font>
During the EDA process, categorical columns with multiple unique values may be identified. To enhance model performance, it is essential to refine these categorical features by grouping values that have low frequency or provide limited predictive information.

Combine categories that occur infrequently or exhibit similar behavior to reduce sparsity and improve model generalisation.

```python
# Combine categories that have low frequency or provide limited predictive information
def combine_low_freq_categories(data, column, threshold=0.05):
    value_counts = data[column].value_counts(normalize=True)
    low_freq = value_counts[value_counts < threshold].index
    data[column] = data[column].apply(lambda x: 'Other' if x in low_freq else x)
    return data

for col in categorical_cols:
    if col in X_train_resampled.columns:
        X_train_resampled = combine_low_freq_categories(X_train_resampled, col)
        X_val = combine_low_freq_categories(X_val, col)
```

```python
# Import libraries or function to plot the ROC curve
from sklearn.metrics import roc_curve, auc

# Define ROC function
fpr, tpr, thresholds = roc_curve(pred_df['Actual'], pred_df['Predicted_Prob'])
roc_auc = auc(fpr, tpr)

```

```python
# Call the ROC function
plt.figure()
plt.plot(fpr, tpr, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.show()  # Display the figure instead of saving it
```

**Sensitivity and Specificity tradeoff**

After analysing the area under the curve of the ROC, check the sensitivity and specificity tradeoff to find the optimal cutoff point.

#### **7.3.2** Predict on training data at various probability cutoffs <font color="red">[1 Mark]</font>

```python
# Create columns with different probability cutoffs to explore the impact of cutoff on model performance
cutoffs = np.arange(0.1, 1.0, 0.1)
results = []
for cutoff in cutoffs:
    pred_df[f'Pred_{cutoff}'] = (pred_df['Predicted_Prob'] >= cutoff).astype(int)
    cm = metrics.confusion_matrix(pred_df['Actual'], pred_df[f'Pred_{cutoff}'])
    TN, FP, FN, TP = cm.ravel()
    results.append({
        'Cutoff': cutoff,
        'Accuracy': metrics.accuracy_score(pred_df['Actual'], pred_df[f'Pred_{cutoff}']),
        'Sensitivity': TP / (TP + FN),
        'Specificity': TN / (TN + FP)
    })
```
